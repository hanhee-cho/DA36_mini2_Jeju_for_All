```python
df = df.dropna(axis = 0)
df.isna().sum()
```

```python
df = df[~df['VISIT_AREA_NM'].str.contains('집')]
df = df[~df['VISIT_AREA_NM'].str.contains('아파트')]
df = df[~df['VISIT_AREA_NM'].str.contains('공항')]
df = df[~df['VISIT_AREA_NM'].str.contains('렌터카')]
df = df[~df['VISIT_AREA_NM'].str.contains('커피')]
df = df[~df['VISIT_AREA_NM'].str.contains('카페')]
df = df[~df['VISIT_AREA_NM'].str.contains('스타벅스')]
df = df[~df['VISIT_AREA_NM'].str.contains('투썸플레이스')]
df = df[~df['VISIT_AREA_NM'].str.contains('빽다방')]
df = df[~df['VISIT_AREA_NM'].str.contains('호텔')]
df = df[~df['VISIT_AREA_NM'].str.contains('리조트')]
df = df[~df['VISIT_AREA_NM'].str.contains('펜션')]
df = df[~df['VISIT_AREA_NM'].str.contains('하우스')]
df = df[~df['VISIT_AREA_NM'].str.contains('식당')]
df = df[~df['VISIT_AREA_NM'].str.contains('마트')]
df = df[~df['VISIT_AREA_NM'].str.contains('GS25')]
df = df[~df['VISIT_AREA_NM'].str.contains('CU')]
df = df[~df['VISIT_AREA_NM'].str.contains('인생네컷')]
df
```
