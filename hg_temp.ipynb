```python
# KeyedVectors 모델 로드
kv_model = KeyedVectors.load("./data/small_fasttext_model.kv")

# 빈 DataFrame 생성 (인덱스와 컬럼은 단어 리스트로 설정)
similarity_df_one = pd.DataFrame(index=unique_values_one, columns=unique_values_one)

# 모든 단어 쌍에 대해 유사도 계산
for word1 in unique_values_one:
    for word2 in unique_values_one:
        # 두 단어가 모두 모델에 있을 때만 유사도 계산
        if word1 in kv_model and word2 in kv_model:
            # 벡터 추출
            vec1 = kv_model[word1].reshape(1, -1)
            vec2 = kv_model[word2].reshape(1, -1)
            # 코사인 유사도 계산
            similarity = cosine_similarity(vec1, vec2)[0][0]
            similarity_df_one.loc[word1, word2] = similarity
        else:
            similarity_df_one.loc[word1, word2] = np.nan  # 모델에 없는 단어는 NaN 처리

display(similarity_df_one)
```

```python
from catboost import Pool
from sklearn.model_selection import train_test_split

X = df[['GENDER', 'AGE_GRP', 'TRAVEL_STYL_5',
       'TRAVEL_STYL_6', 'TRAVEL_STYL_7', 'TRAVEL_STYL_8',
       'TRAVEL_COMPANIONS_NUM']]
y = df['DGSTFN']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape

cat_features = ['GENDER', 'AGE_GRP', 'TRAVEL_STYL_5','TRAVEL_STYL_6',
                'TRAVEL_STYL_7', 'TRAVEL_STYL_8']

# 범주형 데이터 처리를 위한 Pool객체 생성
X_train_pool = Pool(X_train, y_train, cat_features = cat_features)
X_test_pool = Pool(X_test, y_test, cat_features = cat_features)
```

```python
from catboost import CatBoostRegressor
cb_reg = CatBoostRegressor(
    n_estimators = 5000,
    depth = 5,
    learning_rate = .02,
    loss_function = 'RMSE',
    eval_metric = 'RMSE',
    l2_leaf_reg = 10,
    early_stopping_rounds = 200,
    random_state = 0
)
```

```python
cb_reg.fit(X_train_pool, eval_set = X_test_pool, verbose = 200)
```

```python
importances_df = pd.DataFrame({
    'columns' : X_train.columns,
    'importance' : cb_reg.feature_importances_
})
    
importances_df.sort_values(by = 'importance', ascending = False)
```

```python
import unicodedata

# 유니코드 정규화 적용 후 고유 값 확인
df['FCLTY_NM'] = df['FCLTY_NM'].apply(lambda x: unicodedata.normalize('NFKC', str(x)).strip())
places = list(df['FCLTY_NM'].unique())
```

```python
visit_records = X.index

# 빈 DataFrame 생성
dgstfn_pred_df = pd.DataFrame(index=visit_records, columns=places)

dgstfn_pred_df
```

```python
ct_sim
```

```python
for place in places:
    X_train_place = X_train.copy()
    X_train_place['FCLTY_NM'] = place
    
    # Pool 객체 생성
    X_train_pool_place = Pool(X_train_place, cat_features=cat_features + ['FCLTY_NM'])
    
    # 예측 수행
    predicted_scores = cb_reg.predict(X_train_pool_place)
    
    # 예측 결과를 DataFrame에 추가
    dgstfn_pred_df[place] = predicted_scores
    
dgstfn_pred_df
```

```python
# 두 DataFrame의 열 순서를 맞추기 위해 `ct_sim`의 열을 `dgstfn_pred_df`에 맞춤
ct_sim = ct_sim.reindex(columns=dgstfn_pred_df.columns)

# 유사도 기반 점수 계산용 빈 DataFrame 생성
similarity_scores = pd.DataFrame(index=dgstfn_pred_df.index, columns=dgstfn_pred_df.columns)

# 유사도 기반 점수 계산
for record in dgstfn_pred_df.index:
    pred_scores = dgstfn_pred_df.loc[record]
    
    # 유사도 기반 점수 계산: ct_sim과 예측 평점의 점곱(dot product)
    sim_scores = ct_sim.dot(pred_scores)
    similarity_scores.loc[record] = sim_scores

# 가중치 설정
weight_pred = 0.7
weight_sim = 0.3

# 하이브리드 점수 계산
hybrid_scores = (weight_pred * dgstfn_pred_df) + (weight_sim * similarity_scores)
```
