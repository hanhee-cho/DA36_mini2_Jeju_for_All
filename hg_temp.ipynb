```python
# KeyedVectors 모델 로드
kv_model = KeyedVectors.load("./data/small_fasttext_model.kv")

# 빈 DataFrame 생성 (인덱스와 컬럼은 단어 리스트로 설정)
similarity_df_one = pd.DataFrame(index=unique_values_one, columns=unique_values_one)

# 모든 단어 쌍에 대해 유사도 계산
for word1 in unique_values_one:
    for word2 in unique_values_one:
        # 두 단어가 모두 모델에 있을 때만 유사도 계산
        if word1 in kv_model and word2 in kv_model:
            # 벡터 추출
            vec1 = kv_model[word1].reshape(1, -1)
            vec2 = kv_model[word2].reshape(1, -1)
            # 코사인 유사도 계산
            similarity = cosine_similarity(vec1, vec2)[0][0]
            similarity_df_one.loc[word1, word2] = similarity
        else:
            similarity_df_one.loc[word1, word2] = np.nan  # 모델에 없는 단어는 NaN 처리

display(similarity_df_one)
```
