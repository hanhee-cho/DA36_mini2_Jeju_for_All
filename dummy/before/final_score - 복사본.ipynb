```python
def recommend(gender, age_group, comp_num, styl5, styl6, styl7, styl8, topn):

    user_features = pd.DataFrame({
        'GENDER': [gender],
        'AGE_GRP': [age_group],
        'TRAVEL_STYL_5': [styl5],
        'TRAVEL_STYL_6': [styl6],
        'TRAVEL_STYL_7': [styl7],
        'TRAVEL_STYL_8': [styl8],
        'TRAVEL_COMPANIONS_NUM': [comp_num]
    })

    pred_stfn = {}
    
    for place in places:
        user_features_place = user_features.copy()  
        user_features_place['FCLTY_NM'] = place 

        user_pool_place = Pool(user_features_place, cat_features=cat_features + ['FCLTY_NM'])
        
        stfn = cb_reg.predict(user_pool_place)[0]
        stfn=round(stfn,3)
        pred_stfn[place] = stfn

    sorted_recommendations = sorted(pred_stfn.items(), key=lambda x: x[1], reverse=True)[:topn]
    result_df = pd.DataFrame(sorted_recommendations, columns=['FCLTY_NM', 'stfn_pred'])
    
    return result_df
```

```python
import pandas as pd
import unicodedata
from itertools import combinations
from sklearn.metrics.pairwise import cosine_similarity
from gensim.models import KeyedVectors

main_df=pd.read_csv('./data/main_data.csv')
kv_model = KeyedVectors.load("./data/small_fasttext_model.kv")

main_df['CTGRY_TWO_NM'] = main_df['CTGRY_TWO_NM'].apply(lambda x: unicodedata.normalize('NFKC', x))
unique_values_two = main_df['CTGRY_TWO_NM'].unique()
ct_sim = pd.DataFrame(index=unique_values_two, columns=unique_values_two)

# 모든 단어 쌍에 대해 유사도 계산
for word1 in unique_values_two:
    for word2 in unique_values_two:
        # 두 단어가 모두 모델에 있을 때만 유사도 계산
        if word1 in kv_model and word2 in kv_model:
            # 벡터 추출
            vec1 = kv_model[word1].reshape(1, -1)
            vec2 = kv_model[word2].reshape(1, -1)
            # 코사인 유사도 계산
            similarity = cosine_similarity(vec1, vec2)[0][0]
            ct_sim.loc[word1, word2] = similarity
        else:
            ct_sim.loc[word1, word2] = np.nan  # 모델에 없는 단어는 NaN 처리

def place_sim(selected_category, ct_sim, main_df):
    # 선택된 카테고리의 유사도 순으로 정렬
    sim_scores = ct_sim[selected_category].sort_values(ascending=False)
    recommended_places = []  # 추천할 장소 리스트
    category_similarity = {}  # 추천 장소와 유사도 저장할 딕셔너리

    # 유사도가 높은 카테고리부터 장소 추천
    for category in sim_scores.index:  
        facilities = main_df[main_df['CTGRY_TWO_NM'] == category]['FCLTY_NM']

        for fclty in facilities:
            if fclty not in recommended_places:  # 중복된 장소 제외
                recommended_places.append(fclty)
                # 선택된 카테고리와 유사 카테고리의 유사도 저장
                similarity = ct_sim[selected_category][category]
                similarity=round(similarity,3)
                if similarity==1.000:
                    similarity=0.70
                category_similarity[fclty] = similarity
                
    # 추천 장소와 유사도를 데이터프레임으로 변환
    recommendations_df = pd.DataFrame(list(category_similarity.items()), columns=['FCLTY_NM', 'similarity'])
    return recommendations_df
```

```python
def combine_recommendations(gender, age_group, comp_num, styl5, styl6, styl7, styl8, selected_category, ct_sim, main_df):
    reg_df = recommend_2(gender, age_group, comp_num, styl5, styl6, styl7, styl8)
    content_df = place_sim(selected_category, ct_sim, main_df)
    # 유사도에 맞추어 평점 정규화 (1~5) -> (0~1)
    reg_df['stfn_pred_normalized'] = (reg_df['stfn_pred'] - 1) / (5 - 1)
    content_df['similarity_normalized'] = content_df['similarity'] 

    combined_df = pd.merge(reg_df, content_df, on='FCLTY_NM', how='inner')
    combined_df['final_score'] = combined_df['stfn_pred_normalized'] * 0.8 + combined_df['similarity_normalized'] * 0.2
    final_recommendations = combined_df.sort_values(by='final_score', ascending=False).reset_index(drop=True)
    top_10_places = final_recommendations.head(10)
    return top_10_places['FCLTY_NM'].tolist()
    # return final_recommendations
```

```python
# 1. combine_recommendations 함수 정의
def combine_recommendations(gender, age_group, comp_num, styl5, styl6, styl7, styl8, selected_category, ct_sim, main_df):
    reg_df = recommend_2(gender, age_group, comp_num, styl5, styl6, styl7, styl8)
    content_df = place_sim(selected_category, ct_sim, main_df)
    # 평점 정규화 (1~5) -> (0~1)
    reg_df['stfn_pred_normalized'] = (reg_df['stfn_pred'] - 1) / (5 - 1)
    content_df['similarity_normalized'] = content_df['similarity']
    
    combined_df = pd.merge(reg_df, content_df, on='FCLTY_NM', how='inner')
    combined_df['final_score'] = combined_df['stfn_pred_normalized'] * 0.8 + combined_df['similarity_normalized'] * 0.2
    final_recommendations = combined_df.sort_values(by='final_score', ascending=False).reset_index(drop=True)
    
    # top 10 장소 반환
    return final_recommendations.head(10)

# 2. print_recommendations 함수 정의
def print_recommendations(final_recommendations, place_df, selected_options):
    # combine_recommendations 결과와 place_df를 병합
    combined_df = pd.merge(final_recommendations, place_df, on='FCLTY_NM', how='left')
    
    # selected_options 리스트에 포함된 항목들을 필터링 조건으로 적용
    for option in selected_options:
        combined_df = combined_df[combined_df[option] == 1]
    
    # 필터링된 전체 장소명 리스트 반환
    filtered_places = combined_df['FCLTY_NM'].tolist()
    return filtered_places

# 예시 사용
final_recommendations = combine_recommendations(1, 20, 2, 1, 4, 5, 4, '공원', ct_sim, main_df)

# 사용자 선택에 따른 옵션 리스트
selected_options = ['BRLL_GUID_AT', 'WCHAIR_HOLD_AT']  # 예: 점자 안내, 휠체어 보유 옵션만 필터링

filtered_recommendations = print_recommendations(final_recommendations, place_df, selected_options)
filtered_recommendations
```

```python
def print_recommendations(final_recommendations, place_df):

    barrier_free_columns = [
        'BRLL_GUID_AT',          # 점자 안내 여부
        'GUID_DOG_ACP_POSBL_AT', # 안내견 출입 가능 여부
        'DSPSN_TOILET_AT',       # 장애인 화장실 여부
        'WCHAIR_HOLD_AT',        # 휠체어 보유 여부
        'DSPSN_PRKPLCE_AT'       # 장애인 주차장 여부
    ]

    filter_brll_guid_at = False
    filter_guid_dog_acp_posbl_at = False
    filter_dspsn_toilet_at = False
    filter_wchair_hold_at = False
    filter_dspsn_prkplce_at = False
    
    combined_df = pd.merge(final_recommendations, place_df, on='FCLTY_NM', how='left')
    if filter_brll_guid_at:
        for rows in final_recommendations:
            if combined_df['BRLL_GUID_AT'] == 0:
                idx = final_recommendations[final_recommendations['BRLL_GUID_AT']==0].index
                final_recommendations = final_recommendations.drop(idx)
                final_recommendations.reset_index(drop=True, inplace=True)

    if filter_guid_dog_acp_posbl_at:
        for rows in final_recommendations:
            if combined_df['GUID_DOG_ACP_POSBL_AT'] == 0:
                idx = final_recommendations[final_recommendations['GUID_DOG_ACP_POSBL_AT']==0].index
                final_recommendations = final_recommendations.drop(idx)
                final_recommendations.reset_index(drop=True, inplace=True)

    if filter_wchair_hold_at:
        for rows in final_recommendations:
            if combined_df['WCHAIR_HOLD_AT'] == 0:
                idx = final_recommendations[final_recommendations['WCHAIR_HOLD_AT']==0].index
                final_recommendations = final_recommendations.drop(idx)
                final_recommendations.reset_index(drop=True, inplace=True)
                
    if filter_dspsn_toilet_at:
        for rows in final_recommendations:
            if combined_df['DSPSN_TOILET_AT'] == 0:
                idx = final_recommendations[final_recommendations['DSPSN_TOILET_AT']==0].index
                final_recommendations = final_recommendations.drop(idx)
                final_recommendations.reset_index(drop=True, inplace=True)

    if filter_dspsn_prkplce_at:
        for rows in final_recommendations:
            if combined_df['DSPSN_PRKPLCE_AT'] == 0:
                idx = final_recommendations[final_recommendations['DSPSN_PRKPLCE_AT']==0].index
                final_recommendations = final_recommendations.drop(idx)
                final_recommendations.reset_index(drop=True, inplace=True)

    filtered_recommendations = final_recommendations['FCLTY_NM'].head(10)
    return filtered_recommendations
```

```python
def filter_recommendations_by_accessibility(final_recommendations, selected_barrier_free, main_df):
    """
    사용자 선택에 맞는 장애인 편의 시설을 필터링하는 함수.
    
    final_recommendations: combine_recommendations에서 도출된 최종 추천 장소들
    selected_barrier_free: 사용자가 선택한 장애인 편의 시설 리스트
    main_df: 원본 데이터프레임 (시설의 접근성 정보를 포함)
    """
    
    # 각 조건을 필터링할 컬럼 이름에 매핑
    accessibility_mapping = {
        "장애인 주차장": "DSPSN_PRKPLCE_AT",
        "장애인 화장실": "DSPSN_TOILET_AT",
        "휠체어 보유": "WCHAIR_HOLD_AT",
        "안내견 동반 가능": "GUID_DOG_ACP_POSBL_AT",
        "점자 안내": "BRLL_GUID_AT",
        "한국어 음성 안내": "KLANG_VIC_GUID_AT"
    }

    # 필터링할 조건을 설정 (선택된 항목만 True로 설정)
    accessible_conditions = {condition: (condition in selected_barrier_free) 
                             for condition in accessibility_mapping.keys()}
    
    # 필터링된 추천 장소들
    filtered_recommendations = final_recommendations.copy()
    
    for condition, is_selected in accessible_conditions.items():
        if is_selected:  # 조건이 선택된 경우 해당 조건이 1인 장소만 필터링
            column_name = accessibility_mapping[condition]
            filtered_recommendations = filtered_recommendations[filtered_recommendations[column_name] == 1]
    
    return filtered_recommendations


# 사용 예시
# final_recommendations: combine_recommendations 결과
# selected_barrier_free: 사용자가 선택한 장애인 편의 시설 (리스트 형태)

selected_barrier_free = ["장애인 주차장", "휠체어 보유"]  # 예시 선택
final_recommendations = pd.DataFrame({
    'FCLTY_NM': ['장소1', '장소2', '장소3'],
    'final_score': [0.9, 0.8, 0.7],
    'disabled_parking': [1, 0, 1],
    'disabled_restroom': [1, 1, 0],
    'wheelchair_available': [1, 1, 0],
    'guide_dog_allowed': [1, 0, 1],
    'braille_guidance': [0, 0, 1],
    'korean_voice_guidance': [1, 1, 0]
})

# 필터링
filtered_recommendations = filter_recommendations_by_accessibility(final_recommendations, selected_barrier_free, main_df=None)
print(filtered_recommendations)
```

```python
def filter_by_barrier_free_options(final_recommendations, selected_barrier_free, main_df):
    """
    주어진 장소 리스트에서 사용자가 선택한 장애인 편의 시설 옵션을 기준으로 장소를 필터링합니다.

    :param final_recommendations: 최종 추천 장소 리스트 (DataFrame)
    :param selected_barrier_free: 사용자가 선택한 장애인 편의 시설 리스트
    :param main_df: 모든 장소에 대한 장애인 편의 시설 정보가 담긴 DataFrame
    :return: 필터링된 장소 리스트
    """
    
    # 필터링된 장소들
    filtered_places = []

    for place in final_recommendations['FCLTY_NM']:
        # 장소에 대한 장애인 편의 시설 정보를 가져옵니다
        place_facilities = main_df[main_df['FCLTY_NM'] == place]

        # 모든 선택된 장애인 편의 시설 옵션이 해당 장소에서 충족되는지 확인
        matches_all = True
        for option in selected_barrier_free:
            # 각 옵션에 대해 해당 장소가 1(있음)인지 확인
            if place_facilities[option].values[0] == 0:  # 0이면 해당 시설이 없음
                matches_all = False
                break

        if matches_all:
            filtered_places.append(place)

    # 필터링된 장소들을 final_recommendations에서 반환
    filtered_df = final_recommendations[final_recommendations['FCLTY_NM'].isin(filtered_places)]
    return filtered_df

# 예시 사용:
selected_barrier_free = ["DSPSN_PRKPLCE_AT", "DSPSN_TOILET_AT", "WCHAIR_HOLD_AT"]  # 사용자가 선택한 장애인 편의 시설
filtered_recommendations = filter_by_barrier_free_options(final_recommendations, selected_barrier_free, main_df)
```
