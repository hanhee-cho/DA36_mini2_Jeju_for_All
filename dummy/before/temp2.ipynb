```python
from catboost import Pool
from sklearn.model_selection import train_test_split
df['FCLTY_NM'] = df['FCLTY_NM'].astype(str)

X = df[['GENDER', 'AGE_GRP', 'TRAVEL_STYL_5',
       'TRAVEL_STYL_6', 'TRAVEL_STYL_7', 'TRAVEL_STYL_8',
       'TRAVEL_COMPANIONS_NUM','FCLTY_NM']]
y = df['DGSTFN']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.2)

cat_features = ['GENDER', 'AGE_GRP', 'TRAVEL_STYL_5','TRAVEL_STYL_6',
                'TRAVEL_STYL_7', 'TRAVEL_STYL_8', 'FCLTY_NM']

# 범주형 데이터 처리를 위한 Pool객체 생성
X_train_pool = Pool(X_train, y_train, cat_features = cat_features)
X_test_pool = Pool(X_test, y_test, cat_features = cat_features)
```

```python
from catboost import CatBoostRegressor
cb_reg = CatBoostRegressor(
    n_estimators = 5000,
    depth = 5,
    learning_rate = .02,
    loss_function = 'RMSE',
    eval_metric = 'RMSE',
    l2_leaf_reg = 10,
    early_stopping_rounds = 200,
    random_state = 0
)
```

```python
cb_reg.fit(X_train_pool, eval_set = X_test_pool, verbose = 100)
```

```python
importances_df = pd.DataFrame({
    'columns' : X_train.columns,
    'importance' : cb_reg.feature_importances_
})
    
importances_df.sort_values(by = 'importance', ascending = False)
```

```python
def recommend(gender, age_group, comp_num, styl5, styl6, styl7, styl8, topn):

    user_features = pd.DataFrame({
        'GENDER': [gender],
        'AGE_GRP': [age_group],
        'TRAVEL_STYL_5': [styl5],
        'TRAVEL_STYL_6': [styl6],
        'TRAVEL_STYL_7': [styl7],
        'TRAVEL_STYL_8': [styl8],
        'TRAVEL_COMPANIONS_NUM': [comp_num]
    })

    pred_stfn = {}
    
    for place in places:
        user_features_place = user_features.copy()  
        user_features_place['FCLTY_NM'] = place 

        user_pool_place = Pool(user_features_place, cat_features=cat_features + ['FCLTY_NM'])
        
        stfn = cb_reg.predict(user_pool_place)[0]
        pred_stfn[place] = stfn

    sorted_recommendations = sorted(pred_stfn.items(), key=lambda x: x[1], reverse=True)[:topn]
    result_df = pd.DataFrame(sorted_recommendations, columns=['FCLTY_NM', 'stfn_pred'])
    
    return result_df
```

```python
import pandas as pd
# from gensim.models.fasttext import load_facebook_model
from itertools import combinations
from sklearn.metrics.pairwise import cosine_similarity
from gensim.models import KeyedVectors
# import fasttext
```

```python
df = pd.read_csv('./data/main_data.csv')
# 유니코드 정규화 후 고유 값 확인
df['CTGRY_TWO_NM'] = df['CTGRY_TWO_NM'].apply(lambda x: unicodedata.normalize('NFKC', x))
unique_values_two = df['CTGRY_TWO_NM'].unique()
unique_count_two = len(unique_values_two)
```

```python
# 빈 DataFrame 생성 (인덱스와 컬럼은 단어 리스트로 설정)
kv_model = KeyedVectors.load("./data/small_fasttext_model.kv")
ct_sim = pd.DataFrame(index=unique_values_two, columns=unique_values_two)

# 모든 단어 쌍에 대해 유사도 계산
for word1 in unique_values_two:
    for word2 in unique_values_two:
        # 두 단어가 모두 모델에 있을 때만 유사도 계산
        if word1 in kv_model and word2 in kv_model:
            # 벡터 추출
            vec1 = kv_model[word1].reshape(1, -1)
            vec2 = kv_model[word2].reshape(1, -1)
            # 코사인 유사도 계산
            similarity = cosine_similarity(vec1, vec2)[0][0]
            ct_sim.loc[word1, word2] = similarity
        else:
            ct_sim.loc[word1, word2] = np.nan  # 모델에 없는 단어는 NaN 처리

display(ct_sim)
```

```python
catg_similarity = {}

# 모든 장소에 대해 유사도 점수 계산
for _, row in train_df.iterrows():
    fclty_name = row['FCLTY_NM']
    category = row['CTGRY_TWO_NM']

    # 현재 장소의 카테고리와 모든 카테고리 간 유사도를 가져와서 저장
    similarity = ct_sim[category] if category in ct_sim.index else np.nan
    catg_similarity[fclty_name] = similarity

# DataFrame으로 변환하여 반환
similarity_df = pd.DataFrame(catg_similarity).transpose().reset_index()
similarity_df.columns = ['Place'] + list(similarity_df.columns[1:])

similarity_df
```

```python
def category_score(selected_category):
    sim_scores = ct_sim[selected_category].sort_values(ascending=False)
    recommend_fclty = [] 

    for category in sim_scores.index:  
        facilities = train_df[train_df['CTGRY_TWO_NM'] == category]['FCLTY_NM']

        for fclty in facilities:
            if fclty not in recommend_fclty: 
                recommend_fclty.append(fclty)
        if len(recommend_fclty) >= 10: 
            break
    recommend_fclty = recommend_fclty[:10]  
    return recommend_fclty

def get_name_score(recommend_fclty, selected_category):
    catg_similarity = {} 

    for rfc in recommend_fclty:
        category = train_df[train_df['FCLTY_NM'] == rfc]['CTGRY_TWO_NM'].values[0]
        similarity = ct_sim[selected_category][category]  # 선택한 카테고리와 해당 카테고리의 유사도
        catg_similarity[rfc] = similarity
    return catg_similarity

def get_recommendations_and_similarities(selected_category):
    recommended_places = category_score(selected_category)
    category_similarities = get_name_score(recommended_places, selected_category)
    recommendations_df = pd.DataFrame(list(category_similarities.items()), columns=['Place', 'Similarity'])
    return recommendations_df

selected_category = input("Enter category: ")
recommendations_df = get_recommendations_and_similarities(selected_category)

recommendations_df
```
